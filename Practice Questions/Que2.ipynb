```python
# Filter Using NumPy Conditions
# Create a DataFrame with 10 rows and 3 columns filled with random floats between 0 and 1.
# ---Use a NumPy condition to filter all rows where all values are greater than 0.5.
# ---Count how many rows satisfy this condition.
import pandas as pd
import numpy as np
```

```python
# Create DataFrame
columns=["A","B","C"]
data=np.random.rand(10,3)
df=pd.DataFrame(data,columns=columns)
df
```

```python
# (A) Use a NumPy condition to filter all rows where all values are greater than 0.5.
filtered_df = df[df.gt(0.5).all(axis=1)]
filtered_df
```

```python
# Count how many rows satisfy this condition.
count=len(filtered_df)
print(f"Satisfied Rows Count: {count}")
```
