```python
import pandas as pd
import numpy as np

# Q1 Create a DataFrame of Random Integers (5 Rows and 4 Columns['A','B','C','D']) -
# Now:
# Find the mean of each column using NumPy.
# Replace all values less than 50 with NaN using Pandas.
```

```python
Columns = ["A", "B", "C", "D"]
data = np.random.randint(1, 101, size=(5, 4))
df = pd.DataFrame(data, columns=Columns)
df
```

```python
# Now Find mean of each column
for row in df.columns:
  print(f"Mean of Column {row}: {df[row].mean()}")
```

```python
# Now Replace all the values with NaN if the number is less than 50
df2 = df.mask(df < 50)
df2
```
